```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
```

```python
data = load_iris() # sklearn内置了很多数据集
print(type(data))
print(dir(data)) # 查看对象所有的属性和方法
```

```python
help(KNeighborsClassifier()) # 查看一个函数的参数的意义和取值范围
```

```python
X = data['data']
Y = data['target']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
model = KNeighborsClassifier(n_neighbors=3) # 
model.fit(X_train, Y_train) # 训练模型。对于KNN来说，这一步是把训练集的数据存到一个高效的数据结构中，方便查找
```

```python
# 将要预测的数据传入，得到模型的预测结果
Y_pred = model.predict(X_test)
print(Y_pred)
```

### <center>评估预测结果，我们选取acc作为评估指标</center>
$$Accuracy = \frac{TP + TN}{TP + TN + FP + FN}$$

```python
corr_pred = np.count_nonzero(Y_pred == Y_test) # Y_test是我们的ground truth
acc = corr_pred / len(Y_test) # 预测正确的 除以 总数 就是我们的acc指标
print('Accuracy : {:.2f}%'.format(acc * 100))
```
